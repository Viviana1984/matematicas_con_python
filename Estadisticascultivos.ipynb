Imagina que un grupo de amigos quiere saber mas sobre su cosecha de manzanas. Cosecha 10 manzanas y quiere saber y analizar su peso para ver como es su producción este año.
Queremos calcular: (peso promedio de las manzanas)Media, (peso central) Mediana, (Peso más común) Moda, (Qué tanto varían los pesos) Desviación estándar

```python
import numpy as np
import statistics as stats

#Pesos de las manzanas en gramos 🍎
manzanas = [150, 180, 200, 170, 150, 190, 200, 210, 150, 180]

#Estadísticas descriptivas
media= np.mean(manzanas) #promedio
mediana = np.median(manzanas) #Valor central
moda = stats.mode(manzanas) #Valor mas común
desviacion = np.std(manzanas) #Desviación estándar

#resultados:
print(f"Análisis de la cosecha de manzanas 🍎 ")
print(f"Media:{media}g")
print(f"Mediana:{mediana}g")
print(f"Moda:{moda}g")
print(f"Desviación estándar:{desviacion:.2f}g")
```
